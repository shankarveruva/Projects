```python
import pandas as pd
df = pd.read_excel('assignment-01-data-unformated.xlsx', header = 1)
df
```

```python
softCoralsDF = df[['name', 'longitude', 'latitude', 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010]]
softCoralsDF
```

```python
softCoralsDF = pd.melt(softCoralsDF, id_vars =['name', 'longitude', 'latitude'], value_vars =[2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010])
softCoralsDF['coral'] = 'soft corals'
softCoralsDF = softCoralsDF.rename(columns = {'variable': 'year'})
softCoralsDF
```

```python
seaFansDF = df[['name', 'longitude', 'latitude', '2017.1', '2016.1', '2015.1', '2014.1', '2013.1', '2012.1', '2011.1', '2010.1']]
seaFansDF = seaFansDF.rename(columns = {'2017.1': 2017, '2016.1': 2016, '2015.1': 2015, '2014.1': 2014, '2013.1': 2013, '2012.1': 2012, '2011.1': 2011, '2010.1': 2010})
seaFansDF
```

```python
seaFansDF = pd.melt(seaFansDF, id_vars =['name', 'longitude', 'latitude'], value_vars =[2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010])
seaFansDF['coral'] = 'sea fans'
seaFansDF = seaFansDF.rename(columns = {'variable': 'year'})
seaFansDF
```

```python
blueCoralsDF = df[['name', 'longitude', 'latitude', '2017.2', '2016.2', '2015.2', '2014.2', '2013.2', '2012.2', '2011.2', '2010.2']]
blueCoralsDF = blueCoralsDF.rename(columns = {'2017.2': 2017, '2016.2': 2016, '2015.2': 2015, '2014.2': 2014, '2013.2': 2013, '2012.2': 2012, '2011.2': 2011, '2010.2': 2010})
blueCoralsDF
```

```python
blueCoralsDF = pd.melt(blueCoralsDF, id_vars =['name', 'longitude', 'latitude'], value_vars =[2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010])
blueCoralsDF['coral'] = 'blue corals'
blueCoralsDF = blueCoralsDF.rename(columns = {'variable': 'year'})
blueCoralsDF
```

```python
hardCoralsDF = df[['name', 'longitude', 'latitude', '2017.3', '2016.3', '2015.3', '2014.3', '2013.3', '2012.3', '2011.3', '2010.3']]
hardCoralsDF = hardCoralsDF.rename(columns = {'2017.3': 2017, '2016.3': 2016, '2015.3': 2015, '2014.3': 2014, '2013.3': 2013, '2012.3': 2012, '2011.3': 2011, '2010.3': 2010})
hardCoralsDF
```

```python
hardCoralsDF = pd.melt(hardCoralsDF, id_vars =['name', 'longitude', 'latitude'], value_vars =[2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010])
hardCoralsDF['coral'] = 'hard corals'
hardCoralsDF = hardCoralsDF.rename(columns = {'variable': 'year'})
hardCoralsDF
```

```python
seaPensDF = df[['name', 'longitude', 'latitude', '2017.4', '2016.4', '2015.4', '2014.4', '2013.4', '2012.4', '2011.4', '2010.4']]
seaPensDF = seaPensDF.rename(columns = {'2017.4': 2017, '2016.4': 2016, '2015.4': 2015, '2014.4': 2014, '2013.4': 2013, '2012.4': 2012, '2011.4': 2011, '2010.4': 2010})
seaPensDF
```

```python
seaPensDF = pd.melt(seaPensDF, id_vars =['name', 'longitude', 'latitude'], value_vars =[2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010])
seaPensDF['coral'] = 'sea pens'
seaPensDF = seaPensDF.rename(columns = {'variable': 'year'})
seaPensDF
```

```python
softSeaCoralsDF = softCoralsDF.append(seaFansDF, ignore_index = True) 
softSeaBlueCoralsDF = softSeaCoralsDF.append(blueCoralsDF, ignore_index = True)
softSeaBlueHardCoralsDF = softSeaBlueCoralsDF.append(hardCoralsDF, ignore_index = True)
finalDF = softSeaBlueHardCoralsDF.append(seaPensDF, ignore_index = True)
```
